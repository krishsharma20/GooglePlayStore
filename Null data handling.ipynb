# Null value handeling by replacing using SimpleImputer

```python
df3=pd.read_csv("./googleplaystore.csv")
df3.head()
```

# checking the number of null data in every column

```python
df3.isnull().sum()
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')

df3.iloc[:,2:3]= imputer.fit_transform(df3.iloc[:,2:3])

df3.iloc[:,4:5]= imputer.fit_transform(df3.iloc[:,4:5])
```

```python
df3.isnull().sum()
```

So now we have some more null values present in the other columns. So we can see they are very less as compared to the size of the dataset( less than 1%) so we can just remove those null values.

```python
df3.dropna(inplace=True)
```
